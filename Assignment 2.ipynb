```python
df = df.drop(columns=["lights"])
```

```python
X = df.drop(columns="Appliances")
y = df["Appliances"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
#I will be splitting the train and test data to (75,25)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=39)
```

```python
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#To predict on the train set/data, we have

lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
```

```python
train_predictions = lr.predict(X_train_scaled)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
# The Mean Absolute Error is then given as 
mae = mean_absolute_error(y_train, train_predictions)
mae
```

```python
print ('The mean absolute error in 3 decimal places is:', round(mae, 3))
```

```python
rmse_train = np.sqrt(mean_squared_error(y_train, train_predictions))
print ('The root mean absolute error is:', (rmse_train))
```

```python
# To Predict on the test set
from sklearn.metrics import mean_absolute_error
```

```python
test_preds = lr.predict(X_test_scaled)
```

```python
mae_test_set = mean_absolute_error(y_test, test_preds)

print('The Mean Absolute Error for the test set is given as:', (mae_test_set))
```

```python
from sklearn.metrics import mean_squared_error, r2_score

rmse_test_set = np.sqrt(mean_squared_error(y_test, test_preds))
rmse_test_set
```

```python
#To check the data for overfitting
from sklearn.metrics import r2_score

train_r2 = r2_score(y_train, train_predictions)
test_r2 = r2_score(y_test, test_preds)
```
